```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("ImageForOpenCV.jpg")
newimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(img)
plt.show()
plt.imshow(newimg)
plt.show()
```

```python
cv2.imshow("Road",img)
gray=cv2.imread("ImageForOpenCV.jpg",cv2.IMREAD_GRAYSCALE)
cv2.imshow("Road Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
